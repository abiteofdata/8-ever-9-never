```python
import numpy as np
import random
import itertools

#def remove(deck, r):
#    deck.pop(r)
#    deck.insert(r,'e')

def generate_hand(m,n,t):
    
    if (m < 0) or (m > 13) or (n < 0) or (n > 13):
        print('Invalid input!')
        return
    
    m_known = 13-m
    n_known = 13-n

#    suits = 'cdhs'
#    ranks = '23456789TJQKA'

#    deck = list(''.join(card) for card in itertools.product(ranks, suits))

    deck = ['Ac', 'Kc', 'Qc', 'Jc', 'Tc', '9c', '8c', '7c', '6c', '5c', '4c', '3c', '2c', 
            'Ad', 'Kd', 'Qd', 'Jd', 'Td', '9d', '8d', '7d', '6d', '5d', '4d', '3d', '2d', 
            'Ah', 'Kh', 'Qh', 'Jh', 'Th', '9h', '8h', '7h', '6h', '5h', '4h', '3h', '2h', 
            'As', 'Ks', 'Qs', 'Js', 'Ts', '9s', '8s', '7s', '6s', '5s', '4s', '3s', '2s']

    S = ['As', 'Ks', '7s', '6s', '5s']
    N = ['Js', 'Ts', '9s', '8s']
    W = []
    E = []

    #indices = [17,21,25,29,31,35,39,47,51] 
    indices = [48,47,46,45,44,43,42,40,39] #should be in descending order

    for i in indices:
        deck.pop(i)
 
    #add vacant spaces for W
    if m != 13:    
        for i in range(m_known):
            card = deck.pop(0)
            W.append(card)
            
    #add vacant spaces for E
    if n != 13:
        for i in range(n_known):
            card = deck.pop(0)
            E.append(card)
            
    #add S cards, cannot be spades
    while len(S) != 13:
        r = random.randint(0,len(deck)-1)
        if deck[r][1] != 's':
            card = deck.pop(r)
            S.append(card)    
            
    #add N cards, cannot be spades
    while len(N) != 13:
        r = random.randint(0,len(deck)-1)
        if deck[r][1] != 's':
            card = deck.pop(r)
            N.append(card)
            
    #add W cards
    while len(W) != 13:
        r = random.randint(0,len(deck)-1)
        card = deck.pop(r)
        W.append(card)
        
    #add E cards
    while len(E) != 13:
        r = random.randint(0,len(deck)-1)
        card = deck.pop(r)
        E.append(card)
        
    return (S,W,N,E)

def naive_drop_eval(m,n,t,K): #play AK off the top
    success = 0
    for i in range(K):
        (s,w,n,e) = generate_hand(m,n,t)
        w_count = 0
        for i in range(13):
            if W[i][1] == 's':
                w_count += 1
                
        e_count = 4 - w_count
    
        if 'Qs' in W:
            if w_count <= 2:
                success += 1
        elif 'Qs' in E:
            if e_count <= 2:
                success += 1
    
    print(success/K)
    
def good_drop_eval(m,n,t,K): #play AK off the top, unless west shows a void
    success = 0
    for i in range(K):
        (S,W,N,E) = generate_hand(m,n,t)
        w_count = 0
        for i in range(13):
            if w[i][1] == 's':
                w_count += 1
                
        e_count = 4 - w_count
        
        if w_count == 0:
            success += 1
    
        if 'Qs' in W:
            if w_count <= 2:
                success += 1
        elif 'Qs' in E:
            if e_count <= 2:
                success += 1
    
    print(success/K)
    
def naive_finesse_eval(m,n,t,K): #finesse
    success = 0
    for i in range(K):
        (S,W,N,E) = generate_hand(m,n,t)
        if 'Qs' in E:
            success += 1
            
    print(success/N)
    
def good_finesse_eval(m,n,t,K): #Play top first round and attempt 2nd round finesse
    success = 0
    for i in range(K):
        (S,W,N,E) = generate_hand(m,n,t)
        w_count = 0
        for i in range(13):
            if W[i][1] == 's':
                w_count += 1
        if 'Qs' in E:
            success += 1
        elif 'Qs' in W and w_count == 1:  
            success += 1
        
    print(success/K)
    
good_finesse_eval(11,7,13,10000)
```
