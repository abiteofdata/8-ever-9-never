```python
import numpy as np
import random
import itertools

#def remove(deck, r):
#    deck.pop(r)
#    deck.insert(r,'e')

def generate_hand(m,n,t):

    suits = 'cdhs'
    ranks = '23456789TJQKA'

    deck = list(''.join(card) for card in itertools.product(ranks, suits))

    s = ['As', 'Ks', '7s', '6s', '5s']
    n = ['Js', 'Ts', '9s', '8s']
    w = []
    e = []

    #indices = [17,21,25,29,31,35,39,47,51] 
    indices = [51,47,39,35,31,27,23,19,15] #should be in descending order

    for i in indices:
        deck.pop(i)
    
    
    while len(s) != 13:
        r = random.randint(0,len(deck)-1)
        if deck[r][1] != 's':
            card = deck.pop(r)
            s.append(card)
        
    while len(n) != 13:
        r = random.randint(0,len(deck)-1)
        if deck[r][1] != 's':
            card = deck.pop(r)
            n.append(card)
        
    while len(w) != 13:
        r = random.randint(0,len(deck)-1)
        card = deck.pop(r)
        w.append(card)
        
    while len(e) != 13:
        r = random.randint(0,len(deck)-1)
        card = deck.pop(r)
        e.append(card)
        
    return (s,w,n,e)

def naive_drop_eval(m,n,t,N): #play AK off the top
    success = 0
    for i in range(N):
        (s,w,n,e) = generate_hand(m,n,t)
        w_count = 0
        for i in range(13):
            if w[i][1] == 's':
                w_count += 1
                
        e_count = 4 - w_count
    
        if 'Qs' in w:
            if w_count <= 2:
                success += 1
        elif 'Qs' in e:
            if e_count <= 2:
                success += 1
    
    print(success/N)
    
def good_drop_eval(m,n,t,N): #play AK off the top, unless west shows a void
    success = 0
    for i in range(N):
        (s,w,n,e) = generate_hand(m,n,t)
        w_count = 0
        for i in range(13):
            if w[i][1] == 's':
                w_count += 1
                
        e_count = 4 - w_count
        
        if w_count == 0:
            success += 1
    
        if 'Qs' in w:
            if w_count <= 2:
                success += 1
        elif 'Qs' in e:
            if e_count <= 2:
                success += 1
    
    print(success/N)
    
def naive_finesse_eval(m,n,t,N): #finesse
    success = 0
    for i in range(N):
        (s,w,n,e) = generate_hand(m,n,t)
        if 'Qs' in e:
            success += 1
            
    print(success/N)
    
def good_finesse_eval(m,n,t,N): #Play A first round and attempt 2nd round finesse
    success = 0
    for i in range(N):
        (s,w,n,e) = generate_hand(m,n,t)
        w_count = 0
        for i in range(13):
            if w[i][1] == 's':
                w_count += 1
        if 'Qs' in e:
            success += 1
        elif 'Qs' in w and w_count == 1:  
            success += 1
        
    print(success/N)
    
good_drop_eval(0,0,0,100000)
```
